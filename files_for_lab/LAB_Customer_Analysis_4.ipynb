```python
# Do the distributions for different numerical variables look like a normal distribution

data1 = data.copy()

log_lmbda = lambda x: np.log(x) if np.isfinite(np.log(x)) else np.NAN

data1['costumer_lifetime_value'] = data1['customer_lifetime_value'].apply(log_lmbda)

data1['customer_lifetime_value'] = data1['customer_lifetime_value'].fillna(np.mean(data1['customer_lifetime_value']))

sns.distplot(data1['customer_lifetime_value'])
plt.show()
```

```python
log_lmbda = lambda x: np.log(x) if np.isfinite(np.log(x)) else np.NAN

data1['income'] = data1['income'].apply(log_lmbda)

data1['income'] = data1['income'].fillna(np.mean(data1['income']))

sns.distplot(data1['income'])
plt.show()
```

```python
log_lmbda = lambda x: np.log(x) if np.isfinite(np.log(x)) else np.NAN

data1['months_since_last_claim'] = data1['months_since_last_claim'].apply(log_lmbda)

data1['months_since_last_claim'] = data1['months_since_last_claim'].fillna(np.mean(data1['months_since_last_claim']))

sns.distplot(data1['months_since_last_claim'])
plt.show()
```

```python
#For the numerical variables, check the multicollinearity between the features. 
#Please note that we will use the column total_claim_amount later as the target variable.

mask = np.zeros_like(df.corr())

mask[np.triu_indices_from(mask)] = True #hide the repeated half of the matrix

fig, ax = plt.subplots(figsize=(8, 6))
ax = sns.heatmap(df.corr(), mask=mask, annot=True)
plt.show()
```
